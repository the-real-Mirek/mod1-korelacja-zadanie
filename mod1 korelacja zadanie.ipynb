```python
import pandas as pd
import numpy as np
import statistics as st
import math
import matplotlib.pyplot as plt
kraje=np.array(['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13','x14','x15','x16','x17','x18','x19','x20'])
fertility_rate= np.array([1.3,1.95,2.466,1.6,2.952,1.801,1.34,1.34,1.63,2.326,2.436,6.563,1.283,1.43,1.988,1.61,1.921,2.4,1.495,4.705])
life_expectancy_at_birth =np.array([76.84878,81.40732,77.57895,74.16341,68.84907,73.88595,75.26829,76.26098,80.57244,65.46259,67.5482,55.02451,76.2799,82.29024,69.80695,81.40112,82.19756,74.22683,80.12888,45.55095])
```

```python
średnia1=fertility_rate.mean()
suma1=fertility_rate.sum()
mediana1=st.median(fertility_rate)
```

```python
średnia2=life_expectancy_at_birth.mean()
suma2=life_expectancy_at_birth.sum()
mediana2=st.median(life_expectancy_at_birth)
```

```python
wariancja1_suma=st.variance(fertility_rate)
wariancja2_suma=st.variance(life_expectancy_at_birth)
```

```python
odch_stand1_suma=st.stdev(fertility_rate,średnia1)
odch_stand2_suma=st.stdev(life_expectancy_at_birth)
```

```python
print('Fertility rate. ','średnia:',średnia1,'mediana:',mediana1,'odchylenie standardowe:',odch_stand1_suma)
```

```python
print('Life expectancy at birth. ','średnia:',średnia2,'mediana:',mediana2,'odchylenie standardowe:',odch_stand2_suma)
```

```python
kowariancja_suma=np.cov(fertility_rate, life_expectancy_at_birth)
wspł_korelacji=np.corrcoef(fertility_rate, life_expectancy_at_birth)
print('Kowariancja:', kowariancja_suma, 'Współczynnik korelacji:',wspł_korelacji)
```

```python
wariancja1=list()
odch_stand1=list()
for i in range(0,len(fertility_rate)):
    wariancja1.append(((fertility_rate[i]-średnia1)**2)/(len(fertility_rate)-1))
    i+=1
for i in range(0,len(fertility_rate)):
    odch_stand1.append(math.sqrt(wariancja1[i]))
    i+=1

wariancja2=list()
odch_stand2=list()
for i in range(0,len(life_expectancy_at_birth)):
    wariancja2.append(((life_expectancy_at_birth[i]-średnia2)**2)/(len(life_expectancy_at_birth)-1))
    i+=1
for i in range(0,len(life_expectancy_at_birth)):
    odch_stand2.append(math.sqrt(wariancja2[i]))
    i+=1
```

```python
kowariancja=list()
for i in range(0,len(fertility_rate)):
    kowariancja.append(((fertility_rate[i]-średnia1)*(life_expectancy_at_birth[i]-średnia2))/(len(fertility_rate)-1))
    i+=1
```

```python
wariancja1=np.array(wariancja1)
odch_stand1=np.array(odch_stand1)
wariancja2=np.array(wariancja2)
odch_stand2=np.array(odch_stand2)
kowariancja=np.array(kowariancja)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 5), layout='constrained')
axs[0].plot(kraje, fertility_rate)
axs[0].set_title('fertility_rate')
axs[1].plot(kraje, life_expectancy_at_birth)
axs[1].set_title('life_expectancy_at_birth')
```
